### K-Nearest Neighbors 
 Download the following dataset (from Kaggle) with 110.527 medical appointments and 14 
associated variables (characteristics). Create models that can predict whether a patient will show up for the appointment booked or not.

```python
y = patients['No-show']
X = patients.drop('No-show', axis = 1) 
X = X.drop('PatientId', axis = 1)
X = X.drop('AppointmentID', axis = 1)
X = X.drop('Neighbourhood', axis = 1)
X = X.drop('Gender', axis = 1)
X = X.drop('ScheduledDay', axis = 1)
X = X.drop('AppointmentDay', axis = 1)

from sklearn.neighbors import KNeighborsClassifier

#Splitting the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)

#specifiying algorithm
model = KNeighborsClassifier(n_neighbors = 2)

# Fitting the model
model.fit(X_train, y_train)

# training and test scores
print("Training accuracy", model.score(X_train, y_train))
print("Test set acuracy", model.score(X_test, y_test))
```

```python
y = patients['No-show']
X = patients.drop('No-show', axis = 1) 
X = X.drop('PatientId', axis = 1)
X = X.drop('AppointmentID', axis = 1)
X = X.drop('Neighbourhood', axis = 1)
X = X.drop('Gender', axis = 1)
X = X.drop('ScheduledDay', axis = 1)
X = X.drop('AppointmentDay', axis = 1)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)

Training_accuracy = []
Test_accuracy = []

Neighbors = range(1,7)
from sklearn.neighbors import KNeighborsClassifier

for n_neighbors in Neighbors:
    model = KNeighborsClassifier(n_neighbors = n_neighbors)
    model.fit(X_train, y_train)
    Training_accuracy.append(model.score(X_train, y_train))
    Test_accuracy.append(model.score(X_test, y_test))
    print("Test set acuracy", Test_accuracy)

plt.plot(Neighbors, Training_accuracy, label = 'Training_accuracy')
plt.plot(Neighbors, Test_accuracy, label = 'Test_accuracy')
plt.legend()
```

### Discussion

Medical appointments data was mmodelled using KNN to predict whether a patient will show up for the appointment booked or not. The model predicted the model predicted the class correctly in the test data for 78% occurences. 

KNN can be controlled by varying the number neighbors and analyzing the effect.This is essential in determinng the ideal value of k that correctly identifies higher percentage of occurences. From the plot above, the KNN model identified k = 2 neighbors.

### Linear Regression Models, LM
Concrete is the most important material in civil engineering. Its compressive strength is a highly nonlinear function of age and ingredients. These ingredients include cement, blast furnace slag, fly ash, water, superplasticizer, coarse aggregate, and fine aggregate. Use it to predict Concretes compressive strength in megapascals.

```python
concrete = pd.read_csv(r'C:\Users\ARNOLD\Downloads\Jenga school\Concrete_Data.csv')
#print(concrete_data.head(5))
```

```python
concrete.columns = ['Cement','Blast Furnace','Fly Ash','Water','Superplasticizer','Coarse Aggregate',
                    'Fine Aggregate','Age', 'Concrete Compressive Strength']
#print(concrete.head())
```

```python
#scikitlearn helps to shuffle,  split the data and select the classifier.
from sklearn.neighbors import KNeighborsRegressor

y = concrete['Concrete Compressive Strength']
X = concrete.drop('Concrete Compressive Strength', axis = 1) 

#Splitting the data the train_test_split function  extracts 75% of the rows as the training set and 25% forms the test set. 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)

#specifiying algorithm
reg = KNeighborsRegressor(n_neighbors = 3)

# Fitting the model
reg.fit(X_train, y_train)

# training and test scores
print('Training Accuracy: ', format(reg.score(X_train, y_train)))
print('Test Accuracy: ', format(reg.score(X_test, y_test)))
```

```python
y = concrete['Concrete Compressive Strength']
X = concrete.drop('Concrete Compressive Strength', axis = 1) 

#Splitting the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)

Neighbors = range(1,8)
from sklearn.neighbors import KNeighborsRegressor

Training_accuracy = []
Test_accuracy = []

for n_neighbors in Neighbors:
    reg = KNeighborsRegressor(n_neighbors = n_neighbors)
    reg.fit(X_train, y_train)
    Training_accuracy.append(reg.score(X_train, y_train))
    Test_accuracy.append(reg.score(X_test, y_test))
    print("Test set acuracy", Test_accuracy)

plt.plot(Neighbors, Training_accuracy, label = 'Training_accuracy')
plt.plot(Neighbors, Test_accuracy, label = 'Test_accuracy')
plt.legend()
```
